## 1. 검증데이터 만들기 (3360 케이스)

```python
k = pd.read_csv('data/extra/대전_weather.csv')
k
```

```python
#7개 날짜만 뽑아낸기
dates_to_extract = ['2023-05-29', '2023-07-29', '2024-01-29', '2023-09-29', 
                    '2024-03-29', '2023-11-29', '2023-11-26']

k['date_hour'] = pd.to_datetime(k['date_hour'])

# Filter rows matching the specified dates
v_data = k[k['date_hour'].dt.strftime('%Y-%m-%d').isin(dates_to_extract)]

# Ensure that only 24 hours for each date are selected
v_data = v_data.groupby(v_data['date_hour'].dt.strftime('%Y-%m-%d')).head(24)
day_name_mapping = {
    'Monday': 'Monday',
    'Tuesday': 'Tuesday',
    'Wednesday': 'Wednesday',
    'Thursday': 'Thursday',
    'Friday': 'Friday',
    'Saturday': 'Saturday',
    'Sunday': 'Sunday'
}

# Add the '요일' column based on the 'date_hour' column
v_data['요일'] = v_data['date_hour'].dt.day_name().map(day_name_mapping)
v_data
```

```python
v_data.reset_index(inplace = True, drop = True)
```

```python
# Drop the '지점' and '일시' columns
v_data = v_data.drop(columns=['지점', '일시','지점명'])

# Rename the 'date_hour' column to '날짜'
v_data = v_data.rename(columns={'date_hour': '날짜'})

# Reorder the columns so that '날짜' is placed after '지점명' and before '기온(°C)'
reordered_columns = ['날짜', '기온(°C)','요일', '강수량(mm)', 'year', 'month', 'day', 'hour']
v_data = v_data[reordered_columns]
```

## 2. 정확도 산정에 필요한 test 데이터 가공하기

```python
test = pd.read_csv('data/raw/test_taxi_tims_U.csv',encoding='cp949')
test
```

```python
test = test.drop(columns=['차량이름', '하차시간','하차X좌표','하차Y좌표','승차거리(m)','할증여부','요금'])
test
```

```python
# Filter the dataset for the specified 7 dates
dates_to_extract = ['2023-05-29', '2023-07-29', '2024-01-29', '2023-09-29', 
                    '2024-03-29', '2023-11-29', '2023-11-26']

# Convert the 'date_hour' column to datetime for filtering
test['승차시간'] = pd.to_datetime(test['승차시간'])

# Filter rows matching the specified datestest
test = test[test['승차시간'].dt.strftime('%Y-%m-%d').isin(dates_to_extract)]
test
```

```python
test.reset_index(inplace = True, drop = True)
test
```

```python
test['hour'] = pd.to_datetime(test['승차시간']).dt.hour
test
```

```python
def assign_group(row):
    if row['승차요일'] in ['Monday', 'Tuesday', 'Wednesday', 'Thursday']:
        return 'Weekdays'
    elif row['승차요일'] == 'Friday':
        return 'Friday'
    else:
        return 'Weekend'

test['group'] = test.apply(assign_group, axis=1)
test
```

```python
group_counts = test['group'].value_counts()

# Display the result
group_counts
```

```python
# Randomly sample the Weekend and Weekdays groups to match the Friday group size
friday_size = 2959

# Assuming input_data contains the original data with group information
weekday_sample = test[test['group'] == 'Weekdays'].sample(friday_size, random_state=42)
weekend_sample = test[test['group'] == 'Weekend'].sample(friday_size, random_state=42)
friday_data = test[test['group'] == 'Friday']

# Combine the samples into a single dataset
balanced_data = pd.concat([weekday_sample, weekend_sample, friday_data])

# Verify the balanced dataset by counting group sizes
use_vaild= balanced_data['group'].value_counts()

# Display the group counts
use_vaild
```
