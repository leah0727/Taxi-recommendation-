## 기온, 강수량 피처추가

```python
df
```

```python
df['승차시간'] = pd.to_datetime(df['승차시간'], errors='coerce')
df
```

```python
# 시간 문자열에서 시간만 추출하는 방법
df['hour'] = df['승차시간'].dt.hour
df
```

```python
df['year'] = df['승차시간'].dt.year
df['month'] = df['승차시간'].dt.month
df['day'] = df['승차시간'].dt.day
df
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# 요일별 승차량
day_hour = df.groupby(['승차요일', 'hour']).size().unstack()

# 히트맵 그리기
plt.figure(figsize=(12, 6))
sns.heatmap(day_hour, cmap='Blues', annot=True, fmt="d")
plt.title('Hourly Taxi Pickups by Day of the Week')
plt.ylabel('Day of the Week')
plt.xlabel('Hour of Day')
plt.show()
```

```python
df2 = pd.merge(df, weather[['year', 'month', 'day', 'hour', '기온(°C)', '강수량(mm)']], on=['year', 'month', 'day', 'hour'], how='left')
df2
```

```python
# NaN 값을 가진 인스턴스 삭제
df2.dropna(inplace=True)
df2
```

```python
#df2.to_csv('data/train_1_to.csv',index = False)
```

```python

def assign_group(row):
    if row['승차요일'] in ['Monday', 'Tuesday', 'Wednesday', 'Thursday']:
        return 'Weekdays'
    elif row['승차요일'] == 'Friday':
        return 'Friday'
    else:
        return 'Weekend'

df2['group'] = df2.apply(assign_group, axis=1)
df2
```

```python
weekdays_df = df2[df2['group'] == 'Weekdays']
weekdays_df
weekdays_df.to_csv('data/processed/train_주중data1.csv', index=False) 
```

```python
friday_df = df2[df2['group'] == 'Friday']
friday_df
friday_df.to_csv('data/processed/train_금요일data1.csv', index=False) 
```

```python
weekend_df = df2[df2['group'] == 'Weekend']
weekend_df
weekend_df.to_csv('data/processed/train_주말data1.csv', index=False) 
```
